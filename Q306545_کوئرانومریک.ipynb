[سوال کوئرانومریکدر کوئرا](https://quera.org/problemset/306545)




<h1 align="center">جواب سوال</h1>


```python
# solution.py

def queranumeric(order: list[str], words: list[str]) -> list[str]:
    order_index  = {c: i for i,c in enumerate(order)}

    def word_key(word):
       return [order_index.get(ch,len(order)) for ch in word]

    return sorted(words, key=word_key)
```

# sorted(iterable, key=None, reverse=False)

يک تابع داخلی پایتون است که ابتدا دادهٔ ورودی را دریافت می‌کند


سپس در صورت مشخص شدن كليد مرتب سازي 
هر عنصر را به مقداری قابل مقایسه تبدیل می‌کند تا مرتب‌سازی بر اساس آن انجام شود و


 در نهایت اگر 

 
reverse=True


 نتیجه را به‌صورت نزولی برمی‌گرداند؛ خروجی همیشه یک لیست جدید است و دادهٔ اصلی بدون تغییر باقی می‌ماند.
