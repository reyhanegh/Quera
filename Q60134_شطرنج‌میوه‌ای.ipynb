<h1 align="center">جواب سوال</h1>


```python

def fruits(tuple_of_fruits):
    output = {}
    for fruit in tuple_of_fruits:
        is_sphere = fruit['shape']== 'sphere'
        mass_ok =  300 <= fruit['mass'] <= 600
        volumn_ok = 100 <= fruit['volume'] <= 500

        if is_sphere and mass_ok and volumn_ok:
            name = fruit['name']

            if name in output:
                output[name] +=1
            else:
                output[name] = 1
    return output
```

# test:

```python
v = (
    		{'name':'apple', 'shape': 'sphere', 'mass': 350, 'volume': 120},
    		{'name':'mango', 'shape': 'square', 'mass': 150, 'volume': 120}, 
   			{'name':'lemon', 'shape': 'sphere', 'mass': 300, 'volume': 100},
    		{'name':'apple', 'shape': 'sphere', 'mass': 500, 'volume': 250}
		)
solution = {'apple': 2, 'lemon': 1}
print(fruits(v))

result = fruits(v)
if result == solution:
    print("Test passed")
else:
    print("Test failed")	

```
