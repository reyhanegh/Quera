```python
a, b, c = list(map(int, input().split('?')))

out  = []

out = [
    a * b * c,
    a + b + c,
    a * b + c,
    (a + b) * c,
    (b + c )* a,
    b * c + a
]
# input: 14?100?1

print(max(out))
```
