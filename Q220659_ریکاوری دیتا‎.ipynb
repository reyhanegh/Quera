<h2 align="center">magic number</h2>
<p align="right" dir="rtl">
هر فرمت فایل در ابتدای خودش یک سری بایت خاص دارد که به آن  
<br />
<span dir="ltr">“magic number” یا “file signature”</span> می‌گویند.  
<br />
این <span dir="ltr">magic number</span> به سیستم یا برنامه‌ها می‌گوید که این فایل چه نوعی است، بدون اینکه به پسوند فایل نگاه کنند.

<p align="right" dir="rtl">
برای مثال <span dir="ltr">b'\x25\x50\x44\x46'</span>  هدر یا <span dir="ltr">magic number</span> فایل <span dir="ltr">PDF</span>
است
.
</p>




<h1 align="center">جواب سوال</h1>


```python
# solution.py
headers = {
    b'\x89PNG\r\n\x1a\n': 'PNG',
    b'\xff\xd8\xff': 'JPEG',
    b'\x42\x4d': 'BMP',
    b'\x49\x49\x2a\x00': 'TIFF',
    b'\x47\x49\x46\x38': 'GIF',
    b'\x50\x4b\x03\x04': 'ZIP',
    b'\x7fELF': 'ELF',
    b'\x25\x50\x44\x46': 'PDF',
    b'\x49\x44\x33': 'MP3',
    b'\xff\xfb': 'MPEG',
    b'\x00\x00\x01\x00': 'PDDF',
    b'\x00\x01\x00\x00': 'ICO',
}


def data_recovery(data):
	result = []
	for h in headers.keys():
		if h in data:
			result.append(headers[h])
	return result


```

# TEST

```python
print(data_recovery(b'\x89PNG\r\n\x1a\nBMII*\x00\xff\xd8\xff\xff\xd8\xff\x00\x00\x01\x00\x00'))

# answer: ['PNG', 'JPEG', 'BMP', 'TIFF', 'PDDF', 'ICO']
```
