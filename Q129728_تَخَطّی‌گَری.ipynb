<h1 align="center">تَخَطّی‌گَری</h1>
<h2 align="right">میتونید از لینک زیر به آن دسترسی پیدا کنید </h2>


[سوال تخطی گری در کوئرا](https://quera.org/problemset/129728)




````python
input_string = str(input())
array = []
for i in range(len(input_string)):
    if (ord(input_string[i]) - 97) % 2 == 0:
        array.append(input_string[i])
    else:
        array.append(input_string[i].upper())
array.sort(reverse=True)
answer = ' '.join(array)
print(answer)
````

<h1 align="center">جواب سوال</h1>


````python
# [new_item  if condition  for item in iterable]

print(*sorted([a if (ord(a)-97)%2 == 0 else a.upper() for a in input()], reverse=True), sep=' ') 
# input: reyhaneh
````

<div align="center">
