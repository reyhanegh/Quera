<h1 align="center">تخطی گری</h1>
<h2 align="right">میتونید از لینک زیر به آن دسترسی پیدا کنید </h2>


[سوال تخطی گری در کوئرا](https://quera.org/problemset/129728)


<h1 align="center">جواب سوال</h1>


````python
print(*sorted([a if ord(a)%2==1 else a.upper() for a in str(input())], reverse=True), sep=' ')
````

<h2 align="center">توضیحات</h2>


<h2 align="left">List Comprehension</h2>

<p align="right" style="font-size:20px;">
<strong>
یکی از زیباترین و مهم‌ترین قابلیت‌های پایتون است که اجازه می‌دهد یک لیست جدید را با یک خط کد و به شکل بسیار خوانا بسازی
</strong>
</p>

<h3 align="right">ساختار کلی</h3>


```
[new_item  if condition  for item in iterable]
```

<h3 align="right">در اینجا</h3>


```
[a if ord(a)%2==1 else a.upper() for a in input()]
```


<p align="right"><strong>این قسمت هر کاراکتر ورودی را بررسی می‌کند</strong></p>

<div align="center">

| کد مربوطه | عملکرد |
| ------------- |:-------------:|
| ord(a) % 2 == 1    | اگر فرد باشد، همان کاراکتر نگه داشته می‌شود      |
|     a.upper()  |  اگر زوج باشد    |


</div>

<p align="right"><strong>ord(a)  مقدار اسکی  کاراکتر را برمی‌گرداند  </strong></p>
